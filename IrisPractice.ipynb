```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="muted")
```

```python
# Agregamos los datos desde el dataset

df = pd.read_csv("iris.csv")
df.head()
```

```python
# Vemos la información relacionada con el dataframe

df.info()

# Observamos que no existen valores nulos dentro de este, con ello, ya no es necesario
# realizar limpieza de datos dentro del dataframe 
```

```python
# Vemos los estadisticos basicos del dataframe en los valores numéricos

df.describe()
```

```python
# Ahora mostramos graficamente algunos valores del dataset.
columns = df.columns.tolist()[1:-2]

for x in columns:
    sns.distplot(df[x])
    plt.show()
```
